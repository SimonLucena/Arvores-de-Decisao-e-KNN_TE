# Imports

```python
# Realizar todos os imports necessários para a atividade
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn import metrics
import seaborn as sns
import pandas as pd
import numpy as np
```

# Primeira parte

## Importar Dataset

```python
# Inicializar o Dataset 'titanic'
data = sns.load_dataset('titanic')
```

## Observar a relação das features "pclass" com "class"

```python
# Separar e apresentar a relação entre as colunas 'pclass' e 'classa'
colunas = ['pclass','class']
print(data[colunas])
```

## Observar a quantidade de NaN existente na feature "deck"

```python
# Contar e mostrar quantos valores NaN existem na coluna 'deck'
count_nan = data['deck'].isna().sum()
print(count_nan)
```

## Remover as linhas que teham valores NaN

```python
# Excluir linhas com valores NaN
data_limpa = data.dropna()
```

## Transformar todas as features categóricos usando One-Hot Encoder

```python
# Excluir linhas com NaN(resetar a variável data_limpa)
data_limpa = data.dropna()

# Expecificar quais colunas excluir pelo tipo delas
colunas_categoricas = data_limpa.select_dtypes(include=['object', 'category']).columns

# Especificar configurações da transformação
column_transformer = make_column_transformer((OneHotEncoder(), colunas_categoricas), remainder='passthrough')

# Transformar colunas usando do OneHotEncoder pelas configurações de column_transformer (transforma em Array)
data_limpa = column_transformer.fit_transform(data_limpa)

# Resgatar nomes de colunas
columns_names = column_transformer.get_feature_names_out()

# Transformar em tipo Data
data_limpa = pd.DataFrame(data=data_limpa, columns=columns_names)

# Corrigir colunas para float/int
data_limpa = data_limpa.apply(pd.to_numeric, errors='ignore')

# Mostrar tipos das colunas (opcional)
# data_limpa.info()
```

## y = survived

```python
# Separar o TARGET
y = data_limpa['remainder__pclass']
X = data_limpa.drop('remainder__pclass',axis=1)
```

```python
# Separar sets de treino e de teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=None, stratify=y) # 80% treino e 20% teste
```

```python
# Criação de 2 modelos para KNN e 2 modelos para Árvore de Decisão
model1 = DecisionTreeClassifier(criterion="gini")
model2 = DecisionTreeClassifier(criterion="entropy", max_depth=5)
model3 = KNeighborsClassifier(n_neighbors=15, metric='euclidean', algorithm='ball_tree')
model4 = KNeighborsClassifier(n_neighbors=7, metric='minkowski', algorithm='auto')

# Treinamento dos modelos
model1.fit(X_train, y_train)
model2.fit(X_train, y_train)
model3.fit(X_train, y_train)
model4.fit(X_train, y_train)
```

```python
# Resultados com o conjunto de teste
result1 = model1.predict(X_test)
result2 = model2.predict(X_test)
result3 = model3.predict(X_test)
result4 = model4.predict(X_test)

# Respostas
acc1 = metrics.accuracy_score(result1, y_test)
show1 = round(acc1 * 100)
print("DecisionTree gini:            {}%".format(show1))

acc2 = metrics.accuracy_score(result2, y_test)
show2 = round(acc2 * 100)
print("DecisionTree entropy/depth2:  {}%".format(show2))

acc3 = metrics.accuracy_score(result3, y_test)
show3 = round(acc3 * 100)
print("KNN nn15/euclidian/ball_tree: {}%".format(show3))

acc4 = metrics.accuracy_score(result4, y_test)
show4 = round(acc4 * 100)
print("KNN nn7/minkowski/auto:       {}%".format(show4))
```

# Segunda parte

## Importar Dataset

```python
# Importar dados do Dataset por url
url = "https://raw.githubusercontent.com/tmoura/machinelearning/master/transfusion.data"
data = pd.read_csv(url, header=None)
```

## Operação

```python
# Separar o TARGET
y = data[0]
X = data.drop(0,axis=1)
```

```python
# Normalizar coluna X
X_normalizado = MinMaxScaler().fit_transform(X)
X = pd.DataFrame(X_normalizado, columns=X.columns)
```

```python
# Separar sets de treino e de teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=None, stratify=y) # 90% treino e 10% teste
```

```python
# Criação de 2 modelos para KNN e 2 modelos para Árvore de Decisão
model1 = DecisionTreeClassifier(criterion="gini")
model2 = DecisionTreeClassifier(criterion="entropy", max_depth=5)
model3 = KNeighborsClassifier(n_neighbors=15, metric='euclidean', algorithm='ball_tree')
model4 = KNeighborsClassifier(n_neighbors=7, metric='minkowski', algorithm='auto')

# Treinamento dos modelos
model1.fit(X_train, y_train)
model2.fit(X_train, y_train)
model3.fit(X_train, y_train)
model4.fit(X_train, y_train)
```

```python
# Resultados com o conjunto de teste
result1 = model1.predict(X_test)
result2 = model2.predict(X_test)
result3 = model3.predict(X_test)
result4 = model4.predict(X_test)

# Respostas
acc1 = metrics.accuracy_score(result1, y_test)
show1 = round(acc1 * 100)
print("DecisionTree gini:            {}%".format(show1))

acc2 = metrics.accuracy_score(result2, y_test)
show2 = round(acc2 * 100)
print("DecisionTree entropy/depth2:  {}%".format(show2))

acc3 = metrics.accuracy_score(result3, y_test)
show3 = round(acc3 * 100)
print("KNN nn15/euclidian/ball_tree: {}%".format(show3))

acc4 = metrics.accuracy_score(result4, y_test)
show4 = round(acc4 * 100)
print("KNN nn7/minkowski/auto:       {}%".format(show4))
```
